# Match Momentum Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None) 
pd.set_option("display.max_rows", None) 
```

### Step 1: Load the Data

```python
data = pd.read_csv('match_data Ind w vs Ire w 2nd odi.csv')
data.head(2)
```

### Step 2: Filter Data for Each Team

```python
# Filter data for India Women and Ireland Women
india_data = data[data['Batting Team'] == 'India Women']
ireland_data = data[data['Batting Team'] == 'Ireland Women']
```

```python
# Verify the filtered data for India Women
india_data.head(1)
```

```python
# Verify the filtered data for Ireland Women
ireland_data.head(1)
```

### Step 3: Calculate Runs Contributed by Each Batter

```python
# Group data by 'Striker' and calculate total runs
india_partnerships = india_data.groupby('Striker')['Runs'].sum().reset_index()
ireland_partnerships = ireland_data.groupby('Striker')['Runs'].sum().reset_index()

# Rename columns for clarity
india_partnerships.columns = ['Batter', 'Runs Contributed']
ireland_partnerships.columns = ['Batter', 'Runs Contributed']
```

```python
# Display the data for India Women
india_partnerships
```

```python
# Display the data for Ireland Women
ireland_partnerships
```

### Step 4: Plot the Runs Contributed for India Women and Ireland Women

```python
# Plot for India Women
plt.figure(figsize=(12, 6))
sns.barplot(x='Batter', y='Runs Contributed', data=india_partnerships, palette='Blues_d')
plt.title("India Women Partnerships")
plt.xlabel("Batter")
plt.ylabel("Runs Contributed")
plt.xticks(rotation=45)
plt.show()


# Plot for Ireland Women
plt.figure(figsize=(12, 6))
sns.barplot(x='Batter', y='Runs Contributed', data=ireland_partnerships, palette='Greens_d')
plt.title("Ireland Women Partnerships")
plt.xlabel("Batter")
plt.ylabel("Runs Contributed")
plt.xticks(rotation=45)
plt.show()
```

### Step 6: Momentum Analysis

```python
# Calculate runs per over for India Women
india_momentum = india_data.groupby('Over Number')['Runs'].sum().reset_index()
india_momentum.columns = ['Over Number', 'Runs Scored']

# Calculate runs per over for Ireland Women
ireland_momentum = ireland_data.groupby('Over Number')['Runs'].sum().reset_index()
ireland_momentum.columns = ['Over Number', 'Runs Scored']
```

```python
# Display the momentum data for India Women
print(india_momentum)
```

```python
# Display the momentum data for Ireland Women
print(ireland_momentum)
```

### Step 7: Plot Momentum for Each Team

```python
# Plot momentum for India Women
plt.figure(figsize=(12, 6))
plt.plot(india_momentum['Over Number'], india_momentum['Runs Scored'], marker='o', label='India Women')
plt.title("Momentum Analysis - India Women")
plt.xlabel("Over Number")
plt.ylabel("Runs Scored")
plt.legend()
plt.grid()
plt.show()

# Plot momentum for Ireland Women
plt.figure(figsize=(12, 6))
plt.plot(ireland_momentum['Over Number'], ireland_momentum['Runs Scored'], marker='o', color='green', label='Ireland Women')
plt.title("Momentum Analysis - Ireland Women")
plt.xlabel("Over Number")
plt.ylabel("Runs Scored")
plt.legend()
plt.grid()
plt.show()
```

### Step 8: Key Wickets

```python
# Identify overs where wickets fell for India Women
india_wickets = india_data[india_data['Dismissal Type'].notnull()]
india_wickets = india_wickets.groupby('Over Number').size().reset_index(name='Wickets')
print(india_wickets)
```

```python
# Identify overs where wickets fell for Ireland Women
ireland_wickets = ireland_data[ireland_data['Dismissal Type'].notnull()]
ireland_wickets = ireland_wickets.groupby('Over Number').size().reset_index(name='Wickets')
print(ireland_wickets)
```

### Step 9: Overlay Wickets on Momentum Chart

```python
# Plot momentum with wickets for India Women
plt.figure(figsize=(12, 6))
plt.plot(india_momentum['Over Number'], india_momentum['Runs Scored'], marker='o', label='India Women')
plt.scatter(india_wickets['Over Number'], [0] * len(india_wickets), color='red', label='Wickets', zorder=5)
plt.title("Momentum Analysis with Wickets - India Women")
plt.xlabel("Over Number")
plt.ylabel("Runs Scored")
plt.legend()
plt.grid()
plt.show()

# Plot momentum with wickets for Ireland Women
plt.figure(figsize=(12, 6))
plt.plot(ireland_momentum['Over Number'], ireland_momentum['Runs Scored'], marker='o', color='green', label='Ireland Women')
plt.scatter(ireland_wickets['Over Number'], [0] * len(ireland_wickets), color='red', label='Wickets', zorder=5)
plt.title("Momentum Analysis with Wickets - Ireland Women")
plt.xlabel("Over Number")
plt.ylabel("Runs Scored")
plt.legend()
plt.grid()
plt.show()
```
